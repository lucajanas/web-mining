### Preparing the club-level data

```python
# load data for clubs, transfers and final season tables
df_club = pd.read_pickle("./data/df_clubs_overview.pkl")
df_transfer = pd.read_pickle("./data/df_transfer_data.pkl")
df_final_table = pd.read_pickle("./data/df_final_table.pkl")
df_final_table = df_final_table[df_final_table['SEASON'].astype(int) >= 2004]
df_club_c = pd.merge(df_club, df_transfer, on=['CLUB_NAME', 'season'])
df_club_c = df_club_c[df_club_c['season'].astype(int) >= 2004]
```

```python
def convert_market_value(value):
    try:
        # Entferne Kommas und ersetze sie durch Punkte
        value = value.replace(",", ".")
        
        # Millionen
        if "Mio. €" in value:
            value = value.replace("Mio. €", "")
            return float(value) * 1e6  # Multipliziere mit 1 Million

        # Tausend
        elif "Tsd. €" in value:
            value = value.replace("Tsd. €", "")
            return float(value) * 1e3  # Multipliziere mit 1 Tausend

        else:
            return None  # Falls die Konvertierung fehlschlägt, gebe None zurück

    except Exception as e:
        print(f"Konvertierungsfehler: {e}, Wert: {value}")
        return None  # Falls die Konvertierung fehlschlägt, gebe None zurück
```

```python
# convert market value columns to decimals
df_club_c['AVG_MARKET_VALUE'] = df_club_c['AVG_MARKET_VALUE'].apply(convert_market_value)
df_club_c['TOTAL_MARKET_VALUE'] = df_club_c['TOTAL_MARKET_VALUE'].apply(convert_market_value)
```

### Analyzing the relation between tip round results and actual results

```python
df_perc = df_matches.dropna(subset=["WIN_PERC_HOME", "WIN_PERC_AWAY", "REMIS_PERC"])
```

```python
def get_expected_result(row):
    """Based on the percentages given, return whether the users expect the home team to win, a draw or a win of the away team"""
    win_perc_home, win_perc_away, remis_perc = row
    max_value = max(win_perc_home, win_perc_away, remis_perc)

    # Determine the corresponding result
    if max_value == win_perc_home:
        return 'HOME_WIN', max_value
    elif max_value == win_perc_away:
        return 'AWAY_WIN', max_value
    else:
        return 'DRAW', max_value    
```

```python
df_perc[["EXPECTED_RESULT", "RESULT_PERC"]] = df_perc[["WIN_PERC_HOME", "WIN_PERC_AWAY", "REMIS_PERC"]].copy().apply(get_expected_result, axis=1).values.tolist()
```

```python
df_matches[["WINNER", "LOOSER"]] = df_matches[["HOME_TEAM", "AWAY_TEAM", "RESULT"]].apply(get_winning_team, axis=1).values.tolist()
```

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score
```

```python
accuracy_score(df_perc["RESULT"], df_perc["EXPECTED_RESULT"])
```

```python
balanced_accuracy_score(df_perc["RESULT"], df_perc["EXPECTED_RESULT"])
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

class_labels = ["HOME_WIN", "AWAY_WIN", "DRAW"]

# Create a confusion matrix
confusion = confusion_matrix(df_perc["RESULT"], df_perc["EXPECTED_RESULT"], labels=class_labels)
report = classification_report(df_perc["RESULT"], df_perc["EXPECTED_RESULT"])

print("\nClassification Report:")
print(report)
```

```python
# Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion, display_labels=class_labels)
disp.plot(cmap='Blues', values_format='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### Analyzing the influence of referees on the match result
pro Schiedsrichter die Gewinnwahrscheinlichkeit der einzelnen Mannschaften
= Anzahl Siege pro Schiedsrichter pro Team / Anzahl Spiele pro Schiedsrichter pro Team

```python
df_matches
```

```python
def get_winning_team(row):
    """Based on the input, return the name of the team that won the game and the team that lost"""
    home_team, away_team, result = row
    if result == 'HOME_WIN':
        return home_team, away_team
    if result == 'AWAY_WIN':
        return away_team, home_team
    return 'DRAW', 'DRAW'
```

```python
df_matches[["WINNER", "LOOSER"]] = df_matches[["HOME_TEAM", "AWAY_TEAM", "RESULT"]].apply(get_winning_team, axis=1).values.tolist()
```

```python
df_ref = df_matches[df_matches["WINNER"] != "DRAW"]
```

```python
wins_per_ref = df_ref.groupby(["REFEREE", "WINNER"]).size().reset_index()
wins_per_ref.columns = ["REFEREE", "WINNER", "WINS"]
```

```python
# number of home games per referee and team
total_games_per_ref_home = df_ref.groupby(["REFEREE", "HOME_TEAM"]).size().reset_index()
total_games_per_ref_home.columns = ["REFEREE", "CLUB", "MATCHES"]
total_games_per_ref_home = total_games_per_ref_home.set_index(["REFEREE", "CLUB"])
```

```python
# number of away games per referee and team
total_games_per_ref_away = df_ref.groupby(["REFEREE", "AWAY_TEAM"]).size().reset_index()
total_games_per_ref_away.columns = ["REFEREE", "CLUB", "MATCHES"]
total_games_per_ref_away = total_games_per_ref_away.set_index(["REFEREE", "CLUB"])
```

```python
total_games_per_ref = (total_games_per_ref_home + total_games_per_ref_away).reset_index()
total_games_per_ref = total_games_per_ref.dropna(subset=["MATCHES"])
```

```python
set(df_ref.HOME_TEAM.unique()) == set(df_ref.AWAY_TEAM.unique())
```

```python
# calculate the win percentage for each team and each referee

res = []

for ref in df_ref.REFEREE.unique():
    for club in df_ref.HOME_TEAM.unique():
        num_games = total_games_per_ref[(total_games_per_ref["CLUB"] == club) & (total_games_per_ref["REFEREE"] == ref)]["MATCHES"].values
        if len(num_games) == 0:
            continue
        elif len(num_games) > 1:
            raise RuntimeError(f"More than one entry for {ref} and {club} for games!")
        num_games = num_games[0]
        wins = wins_per_ref[(wins_per_ref["WINNER"] == club) & (wins_per_ref["REFEREE"] == ref)]["WINS"].values
        if len(wins) == 0:
            continue
        elif len(wins) > 1:
            raise RuntimeError(f"More than one entry for {ref} and {club} for wins!")
        wins = wins[0]
        win_perc_for_ref = wins / num_games
        res.append((ref, club, num_games, wins, win_perc_for_ref))
        
df_win_perc = pd.DataFrame(res, columns=['REFEREE', 'CLUB', 'NUM_MATCHES', 'NUM_WINS', 'WIN_PERC'])
```

```python
def barplot(df, target, x="REFEREE"):
    average_win_perc = df['WIN_PERC'].mean()
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(data=df, x=x, y="WIN_PERC")
    ax.bar_label(ax.containers[0], fmt='%.2f')
    plt.axhline(y=average_win_perc, color='red', linestyle='--', label='Durchschnittliche Gewinnquote')
    plt.xticks(rotation=70)
    plt.title(f"Siegesquote {target}")
    plt.tight_layout()

    plt.show()
```
